```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import seaborn as sns
from numpy import mean
from numpy import std
```

```python
df = pd.read_csv('house_price.csv')
```

```python
df.head()
```

```python
from scipy.stats import  norm
import numpy as np
rng= np.arange(df.price_per_sqft.min(), df.price_per_sqft.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df.price_per_sqft.mean(),df.price_per_sqft.std()))

plt.hist(df.price_per_sqft,bins = 10, rwidth = 0.8,density = True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
min_tresh,max_tresh = df.price_per_sqft.quantile([0.001, 0.999])
min_tresh,max_tresh
```

```python
df[df.price_per_sqft<min_tresh]
```

```python
df[df.price_per_sqft>max_tresh]
```

```python
df2 = df[(df.price_per_sqft<max_tresh) & (df.price_per_sqft>min_tresh)]
df2.head()
```

# Visualization


```python
from scipy.stats import  norm
import numpy as np
rng= np.arange(df2.price_per_sqft.min(), df2.price_per_sqft.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df2.price_per_sqft.mean(),df2.price_per_sqft.std()))

plt.hist(df2.price_per_sqft,bins = 10, rwidth = 0.8,density = True)
plt.xlabel('Price of DF removed outliers')
plt.ylabel('Count')
plt.show()

```

# percentile, standard deviation and IQR of price

```python
minimum_price = np.min(df2.price)



maximum_price = np.max(df2.price)



mean_price = np.mean(df2.price)



median_price = np.median(df2.price)


std_price = np.std(df2.price)


first_quartile = np.percentile(df2.price, 25)
third_quartile = np.percentile(df2.price, 75)
inter_quartile = third_quartile - first_quartile
```

```python
print("Minimum price: ${:,.2f}".format(minimum_price))
print("Maximum price: ${:,.2f}".format(maximum_price))
print("Mean price: ${:,.2f}".format(mean_price))
print("Median price ${:,.2f}".format(median_price))
print("Standard deviation of prices: ${:,.2f}".format(std_price))
print("First quartile of prices: ${:,.2f}".format(first_quartile))
print("Second quartile of prices: ${:,.2f}".format(third_quartile))
print("Interquartile (IQR) of prices: ${:,.2f}".format(inter_quartile))
```

```python
upper = df2.price_per_sqft.mean()+ 3*df2.price_per_sqft.std()
upper
```

```python
lower = df2.price_per_sqft.mean() - 3*df2.price_per_sqft.std()
lower
```

```python
df2[(df2.price_per_sqft<upper) | (df2.price_per_sqft>lower)]
```

```python
df3 = df2[(df2.price_per_sqft<upper) | (df2.price_per_sqft>lower)]
df3.head()
```

# zScore

```python
df3['zscore'] = (df3.price_per_sqft - df3.price_per_sqft.mean()) / df3.price_per_sqft.std()
df3.head()
```

```python
df3[df3.zscore >1]
```

```python
df3[df3.zscore < -1]
```

```python
df3[(df3.zscore < -1) | (df3.zscore >1)]
```

```python
from scipy.stats import  norm
import numpy as np
rng= np.arange(df3.price_per_sqft.min(), df3.price_per_sqft.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df3.price_per_sqft.mean(),df3.price_per_sqft.std()))

plt.hist(df3.price_per_sqft,bins = 10, rwidth = 0.8,density = True)
plt.xlabel('Prices')
plt.ylabel('Count')
plt.show()
```
